<h3>Importing data</h3>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers , models
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D , Flatten , MaxPooling2D , Dense , Dropout , GlobalAveragePooling2D
from tensorflow.keras import losses , optimizers
from sklearn.metrics import classification_report, confusion_matrix
import itertools
import pickle
import cv2
# System libraries
from pathlib import Path
import PIL
from PIL import UnidentifiedImageError
import os.path
import tkinter
```

<h4><img src="Image/data.png"> Get Data </h4>

```python
AugmentedPath = "AugmentedAlzheimerDataset/"
```

```python
image_dir = Path(AugmentedPath)

# Get filepaths and labels
filepaths = list(image_dir.glob(r'**/*.JPG')) + list(image_dir.glob(r'**/*.jpg')) 
AugmentedLabels = list(map(lambda x: os.path.split(os.path.split(x)[0])[1], filepaths))

# Convert type to Series
filepaths = pd.Series(filepaths, name='AugmentedFilepath').astype(str)
labels = pd.Series(AugmentedLabels, name='Label')

# Concatenate filepaths and labels
AugementedDF = pd.concat([filepaths, labels], axis=1)  # image_df
print(AugementedDF)
```

```python
AugPath = Path("AugmentedAlzheimerDataset").rglob("*.jpg")
for imageP in AugPath:
    try:
        img = PIL.Image.open(imageP)
    except PIL.UnidentifiedImageError:
        print(imageP)
```

```python
# Display 20 picture of the dataset with their labels
random_index = np.random.randint(0 , len(AugementedDF) , 20)
fig,axes = plt.subplots(nrows=4 , ncols=4 , figsize=(7,7) , subplot_kw={'xticks': [], 'yticks': []} )

for i , ax in enumerate(axes.flat):
    ax.imshow(plt.imread(AugementedDF.AugmentedFilepath[random_index[i]]))
    ax.set_title(AugementedDF.Label[random_index[i]])
plt.tight_layout()
plt.show()
```

```python
train_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2) # set validation split
```

```python
train_image = train_datagen.flow_from_directory(AugmentedPath , target_size=(200,200) , batch_size=32 , class_mode='categorical' , subset='training')
validatiom_image = train_datagen.flow_from_directory(AugmentedPath , target_size=(200,200) , batch_size=32 , class_mode='categorical' , subset='validation')
```

```python
train_image
```

<h3> <img src="Image/model.png"> Build CNN Model </h3>

```python
CNNModel = Sequential([
    Conv2D(16 , (3,3) , activation='relu' , input_shape=(200,200,3) ),
    MaxPooling2D(2,2),
    Conv2D(32 , (3,3) , activation='relu'),
    MaxPooling2D(2,2),
    Conv2D(32 , (3,3) , activation='relu'),
    MaxPooling2D(2,2),
    Conv2D(32 , (3,3) , activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.2),
    Dense(4 , activation='softmax')
])
```

```python
CNNModel.compile(loss='categorical_crossentropy' , optimizer=optimizers.Adam() , metrics=['accuracy'] )
```

```python
CNNhistory = CNNModel.fit(train_image , steps_per_epoch=len(train_image) , validation_data= validatiom_image , validation_steps=len(validatiom_image) , epochs=10)
# save the model
pickle.dump(CNNModel, open('CNNModel.pkl', 'wb'))
```

```python
# Plot Accuracy and validation accuarcy
plt.plot(CNNhistory.history['accuracy'])
plt.plot(CNNhistory.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
# Plot loss and val_loss accuarcy
plt.plot(CNNhistory.history['loss'])
plt.plot(CNNhistory.history['val_loss'])
plt.title('model losses')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['loss', 'val_loss'], loc='upper left')
plt.show()
```

```python
# to save model
keras.models.save_model(CNNModel , 'saved cnn model')
```

```python
path = ["OriginalDataset/NonDemented/26 (100).jpg" , 
        "AugmentedAlzheimerDataset/VeryMildDemented/0247b05b-9b4c-4677-9560-4a144e1363cf.jpg",
        "OriginalDataset/ModerateDemented/moderateDem44.jpg"]
```

```python
img = plt.imread(path[2])
```

```python
def TestInput(CNNModelSaved , imgMatrix):
    labels=["MildDemented" , "ModerateDemented" , "NonDemented" , "VeryMildDemented"]
    imgMatrix = cv2.resize(imgMatrix , (200,200))
    imgMatrix = imgMatrix.astype('float32')
    imgMatrix /= 255.0
    imgMatrix = cv2.cvtColor(imgMatrix, cv2.COLOR_BGR2RGB)
    imgMatrix = imgMatrix.reshape((1, 200, 200, 3))
    c = CNNModelSaved.predict(imgMatrix)
    p = np.argmax(c)
    return labels[p]
```

```python
PredictValue = TestInput(CNNModel , img)
print(PredictValue)
```
